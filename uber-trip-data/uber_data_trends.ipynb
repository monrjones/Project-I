```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt

# loading the 2015 data
uberdata="alluberdata_df.csv"

uber_df = pd.read_csv(uberdata)

# setting up the DF
uber_df = uber_df[["Date","Time","Uber_Vendor"]]
uber_df.head()

```

```python
# creating groups

uber_group = uber_df.groupby(["Date"])
uber_count = uber_group.count()
```

```python
# setting up a list for the dates -- could also use the setting a variable = list(df["column"]) 
date_list = uber_df["Date"].unique()

# creating a list for ridecount
ridecount_list = [uber_count.loc[x, "Time"] for x in date_list]

```

```python
# plotting the lists

plt.plot(date_list, ridecount_list, color='g')

# labeling plot
plt.xlabel('Dates')
plt.ylabel('Rides')
plt.title('Uber Rides per Day in NYC')
# displaying plot
plt.show()
```
