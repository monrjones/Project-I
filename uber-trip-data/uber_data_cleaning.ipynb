```python
import pandas as pd
import numpy as np
import csv
from datetime import datetime

# loading the 2015 data
uber2015="uber-raw-data-janjune-15.csv"

uber2015_df = pd.read_csv(uber2015)
uber2015_df.head()
```

```python
# removing unneeded columns
uber2015columns_df = uber2015_df[["Dispatching_base_num", "Pickup_date"]]
uber2015columns_df.head()
```

```python
# splitting the pickup column into date and time
datetime2015 = uber2015columns_df["Pickup_date"].str.split(" ", n = 1, expand = True) 
datetime2015.head()
```

```python
# creating lists for the split data
time2015 = list(datetime2015[1])
date2015 = list(datetime2015[0])
dispatch2015 = list(uber2015columns_df["Dispatching_base_num"])


```

```python

# creating clean new dataframe
listdata2015 = {
    "Date": date2015,
    "Time": time2015,
    "Uber_Vendor": dispatch2015
    
}

uber2015clean_df = pd.DataFrame(listdata2015)

uber2015clean_df.head()
```

```python

# importing 2014 data
uberapr2014="uber-raw-data-apr14.csv"
ubermay2014="uber-raw-data-may14.csv"
uberjun2014="uber-raw-data-jun14.csv"
uberjul2014="uber-raw-data-jul14.csv"
uberaug2014="uber-raw-data-aug14.csv"
ubersep2014="uber-raw-data-sep14.csv"

# reading 2014
uberapr2014_df = pd.read_csv(uberapr2014)
uberapr2014_df.head()
```

```python
ubermay2014_df = pd.read_csv(ubermay2014)
uberjun2014_df = pd.read_csv(uberjun2014)

uberjul2014_df = pd.read_csv(uberjul2014)

uberaug2014_df = pd.read_csv(uberaug2014)

ubersep2014_df = pd.read_csv(ubersep2014)
```

```python
# combining 2014 data

comboaprmay = uberapr2014_df.append(ubermay2014_df, ignore_index=True)
comboaprjun = comboaprmay.append(uberjun2014_df, ignore_index=True)
comboaprjul = comboaprjun.append(uberjul2014_df, ignore_index=True)
comboapraug = comboaprjul.append(uberaug2014_df, ignore_index=True)
uberaprsep2014_df = comboapraug.append(ubersep2014_df, ignore_index=True)
uberaprsep2014_df.head()
```

```python
# removing unneeded columns
uber2014columns_df = uberaprsep2014_df[["Date/Time", "Base"]]
uber2014columns_df.head()
```

```python
#splitting the pickup into date and time
datetime2014 = uber2014columns_df["Date/Time"].str.split(" ", n = 1, expand = True) 
datetime2014.head()
```

```python
# creating lists for the new data

time2014 = list(datetime2014[1])
date2014 = list(datetime2014[0])
dispatch2014 = list(uber2014columns_df["Base"])
```

```python
#fixing the time formatting for 2014 dates

fixeddate2014 = []


for date in date2014:
    oldformat = datetime.strptime(date, "%m/%d/%Y")
    newformat = datetime.strftime(oldformat, "%Y-%m-%d")
    fixeddate2014.append(newformat)
    
print(fixeddate2014[0:3])
        

```

```python
#new dataframe for clean data
listdata2014 = {
    "Date": fixeddate2014,
    "Time": time2014,
    "Uber_Vendor": dispatch2014
    
}

uber2014clean_df = pd.DataFrame(listdata2014)

uber2014clean_df.head()
```

```python
# putting all the data together

alluberdata_df = uber2014clean_df.append(uber2015clean_df, ignore_index=True)
alluberdata_df
```

```python
# saving data
alluberdata_df.to_csv("alluberdata_df.csv")
```
