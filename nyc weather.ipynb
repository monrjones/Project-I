```python
import pandas as pd
```

```python
weather = "New York weather data 10-2012 to 12-2015.csv"
```

```python
weather_df = pd.read_csv(weather)
```

```python
weather_df.head()
```

```python
weather_df.columns
```

```python
hide_kelvin = weather_df[['date', 'time', 'temp', 'weather_description']]
hide_kelvin.head()
```

```python
renamed_df = hide_kelvin.rename(columns={"date":"Date", "time":"Time", "temp":"Temperature (F)", "weather_description":"Weather Description"})
renamed_df.head()
```

```python
precipitation_list = ['light rain', 'moderate rain', 'mist',
       'heavy intensity rain', 'drizzle', 'light intensity drizzle', 'heavy snow', 'heavy intensity drizzle',
       'light rain and snow', 'sleet', 'snow', 'thunderstorm with rain',
       'thunderstorm with heavy rain', 'thunderstorm with light rain',
       'freezing rain', 'thunderstorm', 'SQUALLS',
       'proximity thunderstorm', 'smoke', 'very heavy rain', 'light snow', 'proximity thunderstorm with rain', 'rain and snow']
```

```python
other_weather_list = ['sky is clear', 'Sky is Clear', 'few clouds', 'scattered clouds', 'broken clouds', 'overcast clouds', 'dust', 'haze', 'fog', 'smoke']
```

```python
all_weather = list(renamed_df["Weather Description"])
```

```python
precipitation=[]

for x in all_weather: 
    if x in precipitation_list:
        precipitation.append(1)
    
    elif x in other_weather_list:
        precipitation.append(0)
    
    else:
        precipitation.append(4)
        
print(precipitation)
```

```python
conditions = renamed_df["Weather Description"].unique()
conditions
```

```python
grouped_date = renamed_df.groupby(['Date'])

print(grouped_date)

grouped_date.head()
```

```python
date_summary = grouped_date.mean()
date_summary.head()
```

```python

avgtemp = list(date_summary["Temperature (F)"])
```

```python
grouped_temp = renamed_df.groupby(['Temperature (F)'])

print(grouped_temp)

grouped_temp.head()
```

```python
date_list = list(renamed_df['Date'])
```

```python
date_precipitation = {
    "Date": date_list,
    "Precipitation": precipitation,
    "Conditions": all_weather
}

precipitation_df = pd.DataFrame(date_precipitation)

precipitation_df
```

```python
grouped_precip = precipitation_df.groupby(["Date"])

precip_score = grouped_precip.sum()

precip_score.head()
```

```python
precip_count = grouped_precip.count()
precip_count.head()
```

```python


dateslist = weather_df["date"].unique()
dateslist
```

```python
raw_score = list(precip_score["Precipitation"])
raw_count = list(precip_count["Precipitation"])

precip_score["Normalized Score"] = precip_score["Precipitation"] / precip_count["Precipitation"]

final_score = list(precip_score["Normalized Score"])
```

```python
weather_data = {
    "Date": dateslist,
    "Precipitation Score": final_score,
    "Avg Temp": avgtemp
}

clean_weather_df = pd.DataFrame(weather_data)
clean_weather_df.head()
```

```python
clean_weather_df.to_csv("clean_weather_data.csv")
```
